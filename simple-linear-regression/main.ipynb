```python
import numpy as np

# input : surfaces en mÂ²
X = np.array([30, 45, 60, 80, 100, 120, 150])

# output : prix en euros
y = np.array([120000, 180000, 240000, 320000, 400000, 480000, 600000])
```

```python
import matplotlib.pyplot as plt

plt.plot(X, y, 'o', color='red')

plt.show()
```

```python
import random

slope = random.random()
intercept = random.random()
```

```python
import sklearn
import sklearn.linear_model


model = sklearn.linear_model.LinearRegression()
model.fit(X.reshape(-1, 1), y)
model.score(X.reshape(-1, 1), y)
model.predict([[35]])
```

```python
plt.plot(X, y, 'o', color='red', label="DonnÃ©es rÃ©elles")
X_range = np.linspace(X.min(), X.max(), 100).reshape(-1, 1)
y_pred = model.predict(X_range)
plt.plot(X_range, y_pred, color="blue", label="RÃ©gression linÃ©aire")
plt.xlabel("Surface (mÂ²)")
plt.ylabel("Prix (kâ¬)")
plt.legend()
plt.grid(True)
plt.show()
```
